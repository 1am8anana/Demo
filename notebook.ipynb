*Data cleaning*

```python
# เลือกเฉพาะคอลัมน์ที่ต้องการ fillna
columns_to_fill = ['ปัจจัยสถาบันการศึกษา1', 'ปัจจัยสถาบันการศึกษา2', 'ปัจจัยสถาบันการศึกษา3', 'ปัจจัยสถาบันการศึกษา4',
                   'ปัจจัยสถาบันการศึกษา5', 'ปัจจัยสถาบันการศึกษา6', 'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร1',
                   'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร2', 'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร3',
                   'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร4', 'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร5',
                   'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร6', 'ปัจจัยในการตัดสินใจ1', 'ปัจจัยในการตัดสินใจ2',
                   'ปัจจัยในการตัดสินใจ3', 'ปัจจัยในการตัดสินใจ4', 'ปัจจัยด้านเหตุผลส่วนตัว1',
                   'ปัจจัยด้านเหตุผลส่วนตัว2', 'ปัจจัยด้านเหตุผลส่วนตัว3', 'ปัจจัยด้านเหตุผลส่วนตัว4',
                   'ปัจจัยด้านเหตุผลส่วนตัว5', 'ปัจจัยด้านสื่อและการประชาสัมพันธ์1', 'ปัจจัยด้านสื่อและการประชาสัมพันธ์2',
                   'ปัจจัยด้านสื่อและการประชาสัมพันธ์3', 'ปัจจัยด้านสื่อและการประชาสัมพันธ์4',
                   'ปัจจัยด้านสื่อและการประชาสัมพันธ์5']

# เลือกเฉพาะคอลัมน์ที่ต้องการ fillna และ fillna ด้วยค่าเฉลี่ยของแต่ละคอลัมน์และปัดเศษทศนิยมด้วย round()
rawset[columns_to_fill] = rawset[columns_to_fill].fillna(rawset[columns_to_fill].mean().round())
```

*Features Engineering*

```python
list_for_drop_col = ["ลำดับ", "เพศ", "โรงเรียน", "สำเร็จการศึกษาจาก", "จังหวัด", 
                     "จังหวัดที่อาศัย", "เกรดเฉลี่ยสะสม", "ความคาดหวังที่นักศึกษาคิดว่าจะได้รับ", "ข้อเสนอแนะ", "หลักสูตรที่เข้าศึกษา"]
rawset.drop(columns=list_for_drop_col, axis=1, inplace=True)
```

```python
# convert NaN to 0
rawset['รายได้ครอบครัว'].fillna(0, inplace=True)

# set value levels
salary_mapping = {
    0: 0,
    'ไม่เกิน 10,000': 1,
    '10,001-20,000': 2,
    '20,001-30,000': 3,
    '20,001-30,000 ': 3,
    '30,001 ขึ้นไป': 4
}

# convert salary(object) to value levels
rawset["ระดับรายได้"] = rawset["รายได้ครอบครัว"].map(salary_mapping)
rawset.drop(columns=["รายได้ครอบครัว"], axis=1, inplace=True)
```

```python
rawset["จัดการเรียนการสอนที่ศูนย์แม่ริม"].head()
```

```python
# define method for encode target variable
def label_target(value):
    if 'ยินดี/พร้อม' in str(value):
        return 0
    else: return 1
```

```python
# encode target variable to binary values(0, 1)
rawset['target'] = rawset['จัดการเรียนการสอนที่ศูนย์แม่ริม'].apply(label_target)
```

```python
rawset['ปัจจัยสถาบันการศึกษา(avg)'] = rawset[['ปัจจัยสถาบันการศึกษา1', 'ปัจจัยสถาบันการศึกษา2', 'ปัจจัยสถาบันการศึกษา3', 'ปัจจัยสถาบันการศึกษา4', 'ปัจจัยสถาบันการศึกษา5', 'ปัจจัยสถาบันการศึกษา6']].mean(axis=1)
rawset.drop(columns=['ปัจจัยสถาบันการศึกษา1', 'ปัจจัยสถาบันการศึกษา2', 'ปัจจัยสถาบันการศึกษา3', 'ปัจจัยสถาบันการศึกษา4', 'ปัจจัยสถาบันการศึกษา5', 'ปัจจัยสถาบันการศึกษา6'], inplace=True)

rawset['ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร(avg)'] = rawset[['ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร1', 'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร2', 'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร3', 'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร4', 'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร5', 'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร6']].mean(axis=1)
rawset.drop(columns=['ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร1', 'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร2', 'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร3', 'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร4', 'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร5', 'ปัจจัยด้านโปรแกรมการศึกษา/หลักสูตร6'], inplace=True)

rawset['ปัจจัยในการตัดสินใจ(avg)'] = rawset[['ปัจจัยในการตัดสินใจ1', 'ปัจจัยในการตัดสินใจ2', 'ปัจจัยในการตัดสินใจ3', 'ปัจจัยในการตัดสินใจ4']].mean(axis=1)
rawset.drop(columns=['ปัจจัยในการตัดสินใจ1', 'ปัจจัยในการตัดสินใจ2', 'ปัจจัยในการตัดสินใจ3', 'ปัจจัยในการตัดสินใจ4'], inplace=True)

rawset['ปัจจัยด้านเหตุผลส่วนตัว(avg)'] = rawset[['ปัจจัยด้านเหตุผลส่วนตัว1', 'ปัจจัยด้านเหตุผลส่วนตัว2', 'ปัจจัยด้านเหตุผลส่วนตัว3', 'ปัจจัยด้านเหตุผลส่วนตัว4', 'ปัจจัยด้านเหตุผลส่วนตัว5']].mean(axis=1)
rawset.drop(columns=['ปัจจัยด้านเหตุผลส่วนตัว1', 'ปัจจัยด้านเหตุผลส่วนตัว2', 'ปัจจัยด้านเหตุผลส่วนตัว3', 'ปัจจัยด้านเหตุผลส่วนตัว4', 'ปัจจัยด้านเหตุผลส่วนตัว5'], inplace=True)

rawset['ปัจจัยด้านสื่อและการประชาสัมพันธ์(avg)'] = rawset[['ปัจจัยด้านสื่อและการประชาสัมพันธ์1', 'ปัจจัยด้านสื่อและการประชาสัมพันธ์2', 'ปัจจัยด้านสื่อและการประชาสัมพันธ์3', 'ปัจจัยด้านสื่อและการประชาสัมพันธ์4', 'ปัจจัยด้านสื่อและการประชาสัมพันธ์5']].mean(axis=1)
rawset.drop(columns=['ปัจจัยด้านสื่อและการประชาสัมพันธ์1', 'ปัจจัยด้านสื่อและการประชาสัมพันธ์2', 'ปัจจัยด้านสื่อและการประชาสัมพันธ์3', 'ปัจจัยด้านสื่อและการประชาสัมพันธ์4', 'ปัจจัยด้านสื่อและการประชาสัมพันธ์5'], inplace=True)
```
