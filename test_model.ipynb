KNN with pca

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
correlation_matrix = X.corr()
ax = plt.axes()
sns.heatmap(correlation_matrix, cmap="Greens", ax=ax)
plt.show()
```

```python
eigvalues, eigvectors = np.linalg.eig(correlation_matrix)
print(f'Eigenvalues length: {eigvalues.size}, Original Number of Features: {X.shape[1]}')
```

```python
info_proportion = eigvalues / eigvectors.sum()
info_percents = info_proportion * 100
```

```python
plt.figure()
plt.plot(info_percents, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Axes')
plt.ylabel('Percent of Information Explained')
plt.show()
```

```python
cumulative_information_percents = np.cumsum(info_percents)
```

```python
plt.figure()
plt.plot(cumulative_information_percents, 'ro-', linewidth=2)
```

```python
mean = X.mean(axis=0)
sttd = X.std(axis=0)
data_matrix_standardized = (X - mean) / sttd
```

```python
pca = PCA()
principal_components = pca.fit_transform(data_matrix_standardized)
```

```python
principal_axes_variance_ratios = pca.explained_variance_ratio_
principal_axes_variance_percents = principal_axes_variance_ratios * 100
```

```python
pca = PCA(n_components=3)
principal_components = pca.fit_transform(data_matrix_standardized)
```

```python
principal_components_data = pd.DataFrame({
    'PC1': principal_components[:, 0],
    'PC2': principal_components[:, 1],
    'class': y,
 })

sns.lmplot(x='PC1', y='PC2', data=principal_components_data, hue='class', fit_reg=False)
plt.show()
```

```python
pcs = pd.DataFrame(principal_components)
```

```python
pcs.head()
```

```python
knn = KNeighborsClassifier()
scores_knn = cross_val_score(knn, pcs, y, cv=5)
```

```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors
```

```python
pcs.columns = ['PC1', 'PC2', 'PC3']

# กำหนดจำนวนเพื่อบอกขนาดของตัวเลือก (n_neighbors)
n_neighbors = 5

# สร้างโมเดล KNN
knn = neighbors.KNeighborsClassifier(n_neighbors=n_neighbors)
knn.fit(pcs, y)

# พล็อตข้อมูลที่ถูกสร้าง
plt.figure()
plt.title("KNN decision boundary (n_neighbors = %d)" % (n_neighbors))

# ทำการ slice ข้อมูล pcs แต่ละ component
pcs0, pcs1, pcs2 = pcs['PC1'], pcs['PC2'], pcs['PC3']

# พล็อตข้อมูล pcs ที่สร้างขึ้น
plt.scatter(pcs1, pcs2, c=y, cmap=ListedColormap(['#FF0000', '#00FF00']), edgecolors='k')
plt.xlabel('PCS 1')
plt.ylabel('PCS 2')
plt.show()
```
